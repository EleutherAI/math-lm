# Goals
- Write filtering criteria that allow us to efficiently determine whether an html file contains math notation. 
- Write a modified version of bs4's `get_text()` that accurately outputs math notation into $\LaTeX{}$

The purpose of this notebook is to understand where bs4's default `get_text()` fails.

```python
from bs4.element import *
from bs4 import BeautifulSoup as BS

import os
```

```python
# loads example html files
example_path = "example_html/"
sources = []
for fle in os.listdir(example_path):
    with open(os.path.join(example_path, fle)) as f:
        sources.append(f.read())
```

```python
print(sources[0])
```

```python
math_overflow = BS(sources[0])
planetmath = BS(sources[1])
wikipedia = BS(sources[2])
```

```python
print(math_overflow.get_text())
```

```python
print(planetmath.get_text())
```

```python
print(wikipedia.get_text())
```

Impressions
- bs4 has **a lot** of unnecessary newlines. 
- bs4 seems to handle mathjax really well, but mathml quite poorly. However, when `x-tex` is used, the latex is right there! we just need to tell bs4 what to keep and what to ignore
- Google's heuristics seem bad. In particular, pages that feature MathJax don't necessarily contain the string `MathJax-Element`. Need better heuristics.
